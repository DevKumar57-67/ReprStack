```python
import ast 
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
merged_data.dropna(inplace=True)
```

```python
#genres
#id
#title
#keyword
#overview 
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace("  ",""))
    return L1
```

```python
merged_data['cast']= merged_data['cast'].apply(collapse)
merged_data['crew']=merged_data['crew'].apply(collapse)

```

```python
merged_data['genres']= merged_data['genres'].apply(collapse)
merged_data['keywords']=merged_data['keywords'].apply(collapse)
```

```python
merged_data.head()
```

```python
merged_data['overview']= merged_data['overview'].apply(lambda x:x.split())
```

```python
new = merged_data.drop(columns=['budget','homepage','keywords','original_language','popularity','production_companies','release_date','revenue','spoken_languages','status','tagline','vote_average','vote_count','movie_id','runtime','genres','cast','crew','production_countries'])
```

#           MODEL BUILDING 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
model =cv.fit_transform(['overview']).toarray()
```

```python
model.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity= cosine_similarity(model)
```

```python
similarity[0]
```

```python
new[new['title']=='John Carter'].index[0]
```

```python
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances =similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(new.iloc(i[5]).title)
        print()
    
```
